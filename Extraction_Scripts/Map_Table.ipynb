```python
import os
import json
import pandas as pd
```

```python
# Table for Map_Transaction data for India State-wise

path = "pulse/data/map/transaction/hover/country/india/state/"

Map_trans = {
    "State":[], "Year":[], "Quarter":[],
    "District":[], "Count":[], "Amount":[]
}

for state in os.listdir(path):
    state_path = os.path.join(path, state)

    for year in os.listdir(state_path):
        year_path = os.path.join(state_path, year)

        for file in os.listdir(year_path):
            if file.endswith(".json"):
                file_path = os.path.join(year_path, file)

                with open(file_path, 'r') as f:
                    data = json.load(f)

                try:
                    for i in data['data']['hoverDataList']:
                        Map_trans["State"].append(state)
                        Map_trans["Year"].append(int(year))
                        Map_trans["Quarter"].append(int(file.strip(".json")))
                        Map_trans["District"].append(i['name'])
                        Map_trans["Count"].append(i['metric'][0]['count'])
                        Map_trans["Amount"].append(i['metric'][0]['amount'])
                except:
                    pass

df_Map_trans = pd.DataFrame(Map_trans)
df_Map_trans
```

```python
# Table for Map_User data for India State-wise

path = "pulse/data/map/user/hover/country/india/state/"

Map_user = {
    "State":[], "Year":[], "Quarter":[],
    "District":[], "Registered_Users":[]
}

for state in os.listdir(path):
    state_path = os.path.join(path, state)

    for year in os.listdir(state_path):
        year_path = os.path.join(state_path, year)

        for file in os.listdir(year_path):
            if file.endswith(".json"):
                file_path = os.path.join(year_path, file)

                with open(file_path, 'r') as f:
                    data = json.load(f)

                try:
                    for district, details in data['data']['hoverData'].items():
                        Map_user["State"].append(state)
                        Map_user["Year"].append(int(year))
                        Map_user["Quarter"].append(int(file.strip(".json")))
                        Map_user["District"].append(district)
                        Map_user["Registered_Users"].append(details['registeredUsers'])
                except:
                    pass

df_Map_user = pd.DataFrame(Map_user)
df_Map_user
```
