```python
import os
import json
import pandas as pd
```

```python
# Table for Top_Transaction data for India State-wise

path = "pulse/data/top/transaction/country/india/state/"

Top_trans = {
    "State":[], "Year":[], "Quarter":[],
    "Entity":[], "Entity_Name":[],
    "Count":[], "Amount":[]
}

for state in os.listdir(path):
    state_path = os.path.join(path, state)

    for year in os.listdir(state_path):
        year_path = os.path.join(state_path, year)

        for file in os.listdir(year_path):
            if file.endswith(".json"):
                file_path = os.path.join(year_path, file)

                with open(file_path, 'r') as f:
                    data = json.load(f)

                d = data.get('data', {})

                # STATES (may be null)
                for i in d.get('states') or []:
                    Top_trans["State"].append(state)
                    Top_trans["Year"].append(int(year))
                    Top_trans["Quarter"].append(int(file.strip(".json")))
                    Top_trans["Entity"].append("State")
                    Top_trans["Entity_Name"].append(i.get('entityName'))
                    Top_trans["Count"].append(i.get('metric', {}).get('count'))
                    Top_trans["Amount"].append(i.get('metric', {}).get('amount'))

                # DISTRICTS (MAIN DATA)
                for i in d.get('districts') or []:
                    Top_trans["State"].append(state)
                    Top_trans["Year"].append(int(year))
                    Top_trans["Quarter"].append(int(file.strip(".json")))
                    Top_trans["Entity"].append("District")
                    Top_trans["Entity_Name"].append(i.get('entityName'))
                    Top_trans["Count"].append(i.get('metric', {}).get('count'))
                    Top_trans["Amount"].append(i.get('metric', {}).get('amount'))

                # PINCODES
                for i in d.get('pincodes') or []:
                    Top_trans["State"].append(state)
                    Top_trans["Year"].append(int(year))
                    Top_trans["Quarter"].append(int(file.strip(".json")))
                    Top_trans["Entity"].append("Pincode")
                    Top_trans["Entity_Name"].append(i.get('entityName'))
                    Top_trans["Count"].append(i.get('metric', {}).get('count'))
                    Top_trans["Amount"].append(i.get('metric', {}).get('amount'))

df_Top_trans = pd.DataFrame(Top_trans)
df_Top_trans
```
